# IPv6 Header

```python
from byteclasses.headers.network.ipv6 import IPv6Header
from rich.console import Console
from rich.table import Table
```

```python
ipv6_hdr = IPv6Header()
data = bytearray(b"\x60\x00\x00\x00\x00\x18\x06\x80\x20\x02\x51\x83\x43\x83\x00\x00\x00\x00\x00\x00\x51\x83\x43\x83\x20\x01\x06\x38\x09\x02\x00\x01\x02\x01\x02\xff\xfe\xe2\x75\x96")
ipv6_hdr.attach(memoryview(data))
```

```python
table = Table(title="ipv6 Header")

table.add_column("Name")
table.add_column("Value")

table.add_row("len(ipv6_hdr)", str(len(ipv6_hdr)))
table.add_row("str(ipv6_hdr)", str(ipv6_hdr))
table.add_row("repr(ipv6_hdr)", repr(ipv6_hdr))
table.add_row("ipv6_hdr.vtf", str(ipv6_hdr.vtf))
table.add_row("ipv6_hdr.vtf.version", str(ipv6_hdr.vtf.version))
table.add_row("ipv6_hdr.vtf.traffic_class", str(ipv6_hdr.vtf.traffic_class))
table.add_row("ipv6_hdr.vtf.flow_label", str(ipv6_hdr.vtf.flow_label))
table.add_row("ipv6_hdr.payload_length", str(ipv6_hdr.payload_length))
table.add_row("ipv6_hdr.next_hdr", str(ipv6_hdr.next_hdr))
table.add_row("ipv6_hdr.hop_limit", str(ipv6_hdr.hop_limit))
table.add_row("ipv6_hdr.src_addr", str(ipv6_hdr.src_addr))
table.add_row("ipv6_hdr.src_addr.uint8", str(ipv6_hdr.src_addr.uint8))
table.add_row("ipv6_hdr.src_addr.uint32", str(ipv6_hdr.src_addr.uint32))
table.add_row("ipv6_hdr.src_addr.uint64", str(ipv6_hdr.src_addr.uint64))
table.add_row("ipv6_hdr.dst_addr", str(ipv6_hdr.dst_addr))
table.add_row("ipv6_hdr.dst_addr.uint8", str(ipv6_hdr.dst_addr.uint8))
table.add_row("ipv6_hdr.dst_addr.uint32", str(ipv6_hdr.dst_addr.uint32))
table.add_row("ipv6_hdr.dst_addr.uint64", str(ipv6_hdr.dst_addr.uint64))

console = Console()
console.print(table)
```
